### 这么看太慢，一次来10数字吧 ~

```python
fig, ax = plt.subplots(nrows=2,ncols=5 )
ax = ax.flatten()
for i in range(10):
    img = mnist.test.images[i].reshape(28, 28)
    ax[i].imshow(img, cmap=plt.cm.gray)
    ax[i].set_xticks([])
    ax[i].set_yticks([])
plt.show()
```

### 还是不痛快？那就来25个！

```python
fig, ax = plt.subplots(nrows=5,ncols=5)
ax = ax.flatten()
for i in range(25):
    img = mnist.test.images[i].reshape(28, 28)
    ax[i].imshow(img, cmap=plt.cm.gray)
    ax[i].set_xticks([])
    ax[i].set_yticks([])
plt.show()
```
