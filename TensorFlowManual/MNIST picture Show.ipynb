```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
import cv2
# display one database image
im = mnist.test.images[0].reshape(28, 28)
cv2.imshow("demo", im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

#per row 10 patches, per col 10 patches
BigIm = np.zeros((280, 280))
index = 0
for i in range(10):
    for j in range(10):
        temp = mnist.train.images[index].reshape(28, 28)
        BigIm[28*i:28*(i+1), 28*j:28*(j+1)] = temp
        index += 1

cv2.imshow("demo", BigIm)
cv2.waitKey(0)
```
