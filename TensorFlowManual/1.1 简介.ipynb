# 生成一些三维数据, 然后用一个平面拟合它

```python
import tensorflow as tf
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### 1 使用 NumPy 生成假数据(phony data), 总共 100 个点.

```python
x_data = np.float32(np.random.rand(2, 100)) # 随机输入
y_data = np.dot([0.100, 0.200], x_data) + 0.300 
# np.dot(A, B)：对于二维矩阵，计算真正意义上的矩阵乘积，同线性代数中矩阵乘法的定义。对于一维矩阵，计算两者的内积。
```

### 可视化数据集

```python
x_data = np.float32(np.random.rand(2, 100)) # 随机输入
y_data = np.dot([0.100, 0.200], x_data) + 0.300 
ax = plt.figure().add_subplot(111, projection = '3d') 
#基于ax变量绘制三维图 
#xs表示x方向的变量 
#ys表示y方向的变量 
#zs表示z方向的变量，这三个方向上的变量都可以用list的形式表示 
#m表示点的形式，o是圆形的点，^是三角形（marker) 
#c表示颜色（color for short） 
ax.scatter(x_data[0], x_data[1], y_data, c = 'b', marker = 'o') #点为红色三角形 
  
#设置坐标轴 
ax.set_xlabel('X Label') 
ax.set_ylabel('Y Label') 
ax.set_zlabel('Z Label') 
  
#显示图像 
plt.show()
```

### 构造一个线性模型

```python
b = tf.Variable(tf.zeros([1]))
W = tf.Variable(tf.random_uniform([1, 2], -1.0, 1.0))
y = tf.matmul(W, x_data) + b
```

```python
b
```

```python
W
```

```python
y
```

### 最小化方差

```python
loss = tf.reduce_mean(tf.square(y - y_data))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)
```

### 初始化变量

```python
init = tf.initialize_all_variables()
```

### 启动图 (graph)

```python
sess = tf.Session()
sess.run(init)
```

###  拟合平面

```python
for step in range(0, 201):
    sess.run(train)
    if step % 20 == 0:
        print (step, sess.run(W), sess.run(b))
```

```python
sess.run(W)
```

```python
sess.run(b)
```

### 得到最佳拟合结果 W: [[0.100  0.200]], b: [0.300]

```python
X = [0,sess.run(W)[0][0]]
Y = [0,sess.run(W)[0][1]]
Z = [0,sess.run(b)[0]]

fig = plt.figure()
ax = Axes3D(fig)

Y, X = np.meshgrid(Y, X)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.jet)

plt.show()
```
