### 2、8、16 与 10 进制转换

```python
print(int('1111',2),int('17',8),int('F',16))
print(hex(15),oct(15),bin(15))
```

### ASCII码

```python
print("汉字转为ASCII码：", ord('a'), ord('A'), ord('0'), ord('河'),  ord('北'),  ord('大'),  ord('学'))
print("ASCII码转为汉字：", chr(97),  chr(65),  chr(48),  chr(27827), chr(21271), chr(22823), chr(23398))
```

```python
print(int('6cb3',16),int('5317',16),int('5927',16),int('5b66',16))
print( "%c %c %c %c" % (27827,21271,22823,23398))
```
