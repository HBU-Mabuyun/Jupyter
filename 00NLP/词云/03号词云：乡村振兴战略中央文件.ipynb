```python
import wordcloud
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

# 从外部.txt文件中读取大段文本，存入变量txt中

```python
f = open('words/乡村振兴.txt',encoding='utf-8')
txt = f.read()
print('乡村振兴.txt 字符个数：',len(txt))
print(txt[0:100])
```

### 构建词云对象w，设置词云图片宽、高、字体、背景颜色等参数

```python
w = wordcloud.WordCloud(width=1000,height=700,background_color='white',font_path='msyh.ttc')
```

```python
# 将txt变量传入w的generate()方法，给词云输入文字
w.generate(txt)
```

```python
# 将词云图片导出到当前文件夹
w.to_file('WordClouds/output3-sentence.png')
img=Image.open("wordclouds/output3-sentence.png")
m=np.asarray(img)
plt.imshow(m)
```
