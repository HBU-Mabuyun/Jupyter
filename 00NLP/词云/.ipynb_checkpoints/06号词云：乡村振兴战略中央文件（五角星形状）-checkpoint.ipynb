```python
import wordcloud
import jieba
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

# 导入imageio库中的imread函数，并用这个函数读取本地图片，作为词云形状图片

```python
import imageio
mk = imageio.imread("picture/wujiaoxing.png")
w = wordcloud.WordCloud(mask=mk)
```

# 构建并配置词云对象w，注意要加scale参数，提高清晰度

```python
w = wordcloud.WordCloud(width=1000,
                        height=700,
                        background_color='white',
                        font_path='msyh.ttc',
                        mask=mk,
                        scale=15)
```

```python
# 对来自外部文件的文本进行中文分词，得到string
f = open('words\\乡村振兴.txt',encoding='utf-8')
txt = f.read()
txtlist = jieba.lcut(txt)
string = " ".join(txtlist)

# 将string变量传入w的generate()方法，给词云输入文字
w.generate(string)

# 将词云图片导出到当前文件夹
w.to_file('WordClouds/output6-village.png')
img=Image.open("wordclouds/output6-village.png")
m=np.asarray(img)
plt.imshow(m)
```
