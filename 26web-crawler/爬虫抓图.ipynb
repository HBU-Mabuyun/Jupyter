```python
# -*- coding: UTF-8 -*-
import urllib.request
import re


def get_html(url):
    page = urllib.request.urlopen(url)
    html = page.read()
    return html


def get_img(html):
    html = html.decode('utf-8')
    reg = r'src="(http[^"]+?\.jpg)"'
    img_re = re.compile(reg)
    img_list = re.findall(img_re, html)
    n = 0
    for img_url in img_list:
        urllib.request.urlretrieve(img_url, 'pic/%s.jpg' % n)
        n += 1
    return img_list

ht = get_html("http://tieba.baidu.com/p/5004440579")

for i in get_img(ht):
    print (i)
```
