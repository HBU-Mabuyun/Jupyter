```python
import jieba
%matplotlib inline
import pandas as pd

jieba.load_userdict("dict.txt")

book = "xiongan.txt"
txt = open(book,"r",encoding='utf-8').read()

ex = {'第一节','第二节','第三节'}

ls = []
words = jieba.lcut(txt)
counts = {}
for word in words:
    ls.append(word)
    if len(word) == 1:
        continue
    else:
        counts[word] = counts.get(word,0)+1

for word in ex:
    del(counts[word])

    
items = list(counts.items())
items.sort(key = lambda x:x[1], reverse = True)
for i in range(500):
    word , count = items[i]
    print ("{:<10}{:>5}".format(word,count))

wz = open('ms.txt','w+')
wz.write(str(ls))
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = open('xiongan.txt', 'r',encoding = 'utf-8').read()
word_jieba = jieba.cut(text, cut_all=True)
word_split = " ".join(word_jieba)

my_wordcloud = WordCloud(background_color="white", max_words=500, 
                         max_font_size=50, random_state=30,
                         font_path='C:/Windows/Fonts/simhei.ttf').generate(word_split)

plt.figure(figsize=(9,15)) 
plt.imshow(my_wordcloud)
plt.axis("off")
plt.show()
```
