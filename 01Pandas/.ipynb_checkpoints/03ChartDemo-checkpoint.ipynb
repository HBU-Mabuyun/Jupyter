```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#https://www.yiibai.com/pandas/python_pandas_visualization.html

df = pd.DataFrame(np.random.rand(10,4),columns=['First','Second','Third','Fourth'])
df.plot.bar()
df.plot.bar(stacked=True)
df.plot.barh(stacked=True)
df.plot.area()
df.plot.scatter(x='First', y='Second')
df = pd.DataFrame(3 * np.random.rand(4), index=['First', 'Second', 'Third', 'Fourth'], columns=['x'])
df.plot.pie(subplots=True)
plt.show()
```
