# 一元线性回归

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font = FontProperties(fname = r"c:\windows\fonts\msyh.ttc",size = 10)
```

```python
def runplt():
    plt.figure()
    plt.title('披萨价格与直径数据',fontproperties=font)
    plt.xlabel('直径（英寸）',fontproperties=font)
    plt.ylabel('价格（美元）',fontproperties=font)
    plt.axis([0,25,0,25])
    plt.grid(True)
    return plt
```

```python
plt = runplt()
X = [[6],[8],[10],[14],[18]]
y = [[7],[9],[13],[17.5],[18]]
plt.plot(X,y,'k.')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
model.predict([12][0])
```

```python
print ('Price of 12 inch pizza: %.2f' % model.predict([12][0]))
```

LinearRegression 类是一个估计器（estimator）。 

估计器依据观测值来预测结果。 

在 scikit-learn 里面，所有的估计器都带有 fit() 和 predict() 方法。 

fit() 用来分析模型参数，predict() 是通过 fit() 算出的模型参数构成的模型，对解释变量进行预测获得的值。


```python
plt = runplt()
plt.plot(X,y,'k.')
X2 = [[0],[10],[14],[25]]
model = LinearRegression()
model.fit(X,y)
y2 = model.predict(X2)
plt.plot(X,y,'k.')
plt.plot(X2,y2,'g-')
plt.show()
```

```python
plt = runplt()
plt.plot(X,y,'k.')
y3 = [14.25,14.25,14.25,14.25]
y4 = y2 * 0.5 +5
model.fit(X[1:-1],y[1:-1])
y5 = model.predict(X2)
plt.plot(X,y,'k.')
l1, = plt.plot(X2,y2,'g-')
l2, = plt.plot(X2,y3,'r-.')
l3, = plt.plot(X2,y4,'y-.')
l4, = plt.plot(X2,y5,'o-.')
plt.legend(handles = [l1,l2,l3,l4,], labels = ['y2','y3','y4','y5'], loc = 'best') 
plt.show()
```

```python
plt = runplt()
plt.plot(X,y,'k.')
X2 = [[0],[10],[14],[25]]
model = LinearRegression()
model.fit(X,y)
y2 = model.predict(X2)
plt.plot(X,y,'k.')
plt.plot(X2,y2,'g-')

yr = model.predict(X)
for idx,x in enumerate(X):
    plt.plot([x,x],[y[idx],yr[idx]],'r-')

plt.show()
```

## 残差平方和（residual sum of squares）SSE

```python
import numpy as np
np.mean((model.predict(X) - y) ** 2)
```
