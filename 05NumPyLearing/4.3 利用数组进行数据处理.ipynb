```python
import matplotlib.pyplot as plt #https://zhuanlan.zhihu.com/p/29663486
%matplotlib inline
m, n = (5, 3)
x = np.linspace(0, 1, m)
y = np.linspace(0, 1, n)
X, Y = np.meshgrid(x,y)
plt.plot(X, Y, marker='.', color='blue', linestyle='none')
plt.show()
```
