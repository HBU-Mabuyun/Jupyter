```python
from pyecharts import Map

value = [139, 173, 732, 630, 867]
attr = ['保定市', '邯郸市', '沧州市', '石家庄市', '邢台市']
map = Map("河北地图示例", width=1000, height=600)
map.add("", attr, value, maptype='河北',is_label_show=True, is_visualmap=True, visual_text_color='#000',visual_range=[139,867],
       is_map_symbol_show = True)
map
```
