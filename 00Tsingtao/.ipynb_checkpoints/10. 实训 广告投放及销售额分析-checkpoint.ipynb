# 读入数据

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
data = pd.read_csv("data/Advertising.csv",header=0)
data.head()
```

```python
pd.__version__
```

# 数据理解

```python
sns.pairplot(data,x_vars=['TV','radio','newspaper'],y_vars='sales',height=7,aspect=0.8,kind='reg')
plt.show()
```

# 数据准备

```python
Data=data.drop(['Number','sales'],axis=1)
Data.head()
```

```python
sales=data['sales']
type(sales)
```

```python
import numpy as np
sales=np.ravel(sales)
type(sales)
```

# 模型构建

```python
import statsmodels.api as sm
X_add_const=sm.add_constant(Data.to_numpy())
myModel=sm.OLS(sales,X_add_const)
results=myModel.fit()
print(results.summary())
```

```python
X_add_const
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(Data,sales,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
model=linreg.fit(X_train,y_train)
print(model)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.get_params
```

```python
model.score(X_test,y_test)
```

# 模型预测

```python
y_pred=linreg.predict(X_test)
y_pred
```

# 模型评价

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(range(len(y_pred)),y_pred,'b',label="pred")
plt.plot(range(len(y_pred)),y_test,'r',label="test")
plt.legend(loc="upper right")
plt.xlabel("Location")
plt.ylabel("Money")
plt.show()
```
