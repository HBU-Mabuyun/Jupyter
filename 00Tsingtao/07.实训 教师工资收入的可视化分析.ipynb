```python
import pandas as pd
salaries=pd.read_csv('data/salaries.csv',index_col=0)
```

```python
import seaborn as sns
```

```python
sns.set_style('darkgrid')
sns.set(font='STXihei')
```

```python
sns.stripplot(data=salaries,x='rank',y='salary',jitter=True,alpha=0.5)
```
